<h1>Cleaning and Handling Data</h1>

```python
# Numerical Data
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randint(0, 1000, size=(10, 5)), columns=list('ABCDE'))
df
```

```python
# Categorical Data
df_categorical = pd.DataFrame({"Color": ["Red", "Blue", "Purple", "Yellow"], "Car Brand": ["Dodge", "Mercedes", "Fiat", "Daewoo"]})
df_categorical
```

<h1>Missing Values</h1>

NaNs

```python
# Handle NaNs
create_nan = {'example': [np.nan, np.nan, 20, 30, 40, 50, 60, np.nan, np.nan, 1, 2, 3, 4, 5, np.nan, np.nan, np.nan]}
df_nan = pd.DataFrame(create_nan)
df_nan
```

```python
df_nan.isnull().values.any()
```

```python
df_nan['example'].isnull().values.any()
```

```python
df_nan.isnull().sum()
```

```python
df_clean = df_nan.fillna(0)
df_clean
```

```python
df_nan.fillna(method='ffill')
```

```python
df_nan.fillna(method='bfill')
```

```python
df_drop = df_nan.dropna()
df_drop
```

```python
df_drop.describe()
```
